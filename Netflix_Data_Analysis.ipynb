```python
type_counts = data_frame['type'].value_counts()

plt.figure(figsize=(8, 6))
type_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Répartition du type d'oeuvres dans le dataset")
plt.xlabel('Type')
plt.ylabel("Nombre d'oeuvres")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
country_counts = data_frame['country'].value_counts().head(20)

plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='skyblue')
plt.title('Répartition des oeuvres par pays dans le dataset (Top 20)')
plt.xlabel('Pays')
plt.ylabel("Nombre d'oeuvres")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
year_counts = data_frame['release_year'].value_counts().sort_index().tail(30)

plt.figure(figsize=(10, 6))
year_counts.plot(kind='bar', color='skyblue')
plt.title('La répartition des années du dataset, (Top 30)')
plt.xlabel('release_year')
plt.ylabel("Nombre d'oeuvres")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
year_counts = data_frame['rating'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
year_counts.plot(kind='bar', color='skyblue')
plt.title('La répartition des ratings du dataset')
plt.xlabel('rating')
plt.ylabel("Nombre d'oeuvres")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
movies = data_frame[data_frame['type'] == 'Movie']
movies_duration = movies["duration"].value_counts()

plt.bar(movies_duration.index, movies_duration)
plt.xlabel('durée du film')
plt.ylabel("Nombre d'oeuvres")
plt.title('La répartition de la durée des films du dataset')
plt.show()
```

```python
tv_show = data_frame[data_frame['type'] == 'TV Show']
tv_show_duration = tv_show["duration"].value_counts()

plt.bar(tv_show_duration.index, tv_show_duration)
plt.xlabel('durée des séries')
plt.ylabel("Nombre d'oeuvres")
plt.title('La répartition de la durée des séries du dataset')
plt.show()
```

```python
number_genre_of_series = data_frame["listed_in"].explode().str.strip().value_counts()

plt.figure(figsize=(10, 6))
number_genre_of_series.plot(kind='bar', color='skyblue')
plt.bar(number_genre_of_series.index, number_genre_of_series)
plt.xlabel("Genre d'oeuvres")
plt.ylabel("Nombre d'oeuvres")
plt.title("La répartition des genres d'oeuvres du dataset")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
series_data = data_frame[data_frame["type"] == "TV Show"]
top_5_series = series_data.sort_values(by='duration', ascending=False).head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_5_series["title"], top_5_series["duration"])
plt.xlabel("Nom de la série")
plt.ylabel("Durée de la série")
plt.title("Le top 5 des séries les plus longues")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
movies = data_frame[data_frame["type"] == "Movie"]
top_5_movies = movies.sort_values(by='duration', ascending=False).head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_5_movies["title"], top_5_movies["duration"])
plt.xlabel("Nom du film")
plt.ylabel("Durée du film")
plt.title("Le top 5 des films les plus longs")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
french_works = data_frame[data_frame['country'] == 'France']
director_counts = french_works['director'].value_counts()

plt.figure(figsize=(20, 6))
plt.bar(director_counts.index, director_counts)
plt.xlabel("Nom du directors")
plt.ylabel("nombre d'oeuvres")
plt.title("La répartition des directors des oeuvres françaises")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
data_frame['month_added'] = data_frame['date_added'].dt.to_period('M')
added_counts = data_frame['month_added'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(added_counts.index.to_timestamp(), added_counts.values, color='skyblue', marker='o', linestyle='-')
plt.title("La répartition des oeuvres en fonction de la date d'ajout au catalogue")
plt.xlabel("Date d'ajout")
plt.ylabel("Nombre d'oeuvres ajoutées")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
