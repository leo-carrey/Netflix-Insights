```python
data_frame['date_added'] = pd.to_datetime(data_frame['date_added'])
```

```python
films = data_frame[data_frame['type'] == 'Movie']
films.sort_values(by='date_added').tail(1)[['title','date_added']]
```

```python
series_tv = data_frame[data_frame['type'] == 'TV Show']
series_tv.dropna(subset=['date_added']).sort_values(by='date_added').tail(1)[['title','date_added']]
```

```python
data_frame['duration'] = data_frame['duration'].str.extract('(\d+)')
data_frame['duration'] = pd.to_numeric(data_frame['duration'])
print(data_frame['duration'].dtype)
data_frame.head()
```

```python
data_frame["listed_in"] = data_frame["listed_in"].str.split(', ')
data_frame.head()
```
