### 23. **Visualisez vos données à l’aide des différentes librairies de Matplotlib, Seaborn ou Plotly**

#### a. La répartition du type d'oeuvres du dataset

```python
plt.pie(x=df["type"].value_counts(), labels=df["type"].unique(), autopct="%.2f%%")
plt.legend(df["type"].unique())
plt.title("Répartition du type d'oeuvres")
plt.show()
```

On peut observer qu'il y a **un peu plus de 2x plus** (2.29 x) de films que de séries dans le dataset

#### b. La répartition des oeuvres en fonction des pays du dataset

```python
# On regroupe par pays
countries_count = df["country"].str.split(", ").explode().value_counts()

# On récupère le top 20 des pays (pour la lisibilité)
top_20_countries = countries_count.head(20)

# On affiche
plt.bar(top_20_countries.index, top_20_countries)
plt.xlabel("Pays")
plt.ylabel("Nombre d'oeuvres")
plt.title("Répartition du nombre d'oeuvres par pays")
plt.xticks(rotation=45, ha="right")
plt.show()
```

On voit que **les Etats Unis on produit plus la majorité des films** (plus de 3500 sur 8807), suivi par l'Inde (un peu plus de 1000).

```python
# On affiche
plt.pie(x=countries_count, labels=countries_count.index, autopct="%.2f%%")
plt.title("Répartition des oeuvres en fonction du pays")

plt.show()
```

Le graphique n'est pas tres lisible sur la partie en bas a droite, mais on voit bien que plus d'un tiers des films ont été produits aus Etats Unis, suivi par l'Inde et le Royaume Unis. Plus de 50% du contenu est proposé par ces 3 pays dont plus du tiers (36.84%) pour les Etats Unis.

```python
plt.figure(figsize=(25, 10))
plt.yscale("log")  # Echelle logarihmique (pour la lisibilité)
plt.bar(countries_count.index, countries_count)
plt.title("Répartition des oeuvres en fonction du pays")
plt.xticks(rotation=90)
plt.xlabel("Pays")
plt.ylabel("Echelle logatithme des durées")
plt.show()
```

On constate la même chose ...

#### c. La répartition des années du dataset

```python
# On récupère les années de sortie
release_year_counts = df["release_year"].value_counts()

# On affiche
plt.figure(figsize=(20, 10))
plt.yscale("log")  # Echelle logarithmique (pour la lisibilité)
plt.bar(release_year_counts.index, release_year_counts)
plt.xlabel("Année de sortie")
plt.ylabel("Echelle logarithmique du nombre d'oeuvres")
plt.title("Répartition du nombre d'oeuvres par année de sortie")
plt.xticks(release_year_counts.index, rotation=60)

plt.show()
```

On remarque qu'il y a une **majoritée d'oeuvres qui sont sorties autours de l'année 2018**.

Ceci s'explique en partie car **Netflix a considérablement augmenté son budget de production et de distribution de contenu original au cours de cette période.**

#### d. La répartition des ratings du dataset

```python
# On récupère les ratings
rating_counts = df["rating"].value_counts()

# On affiche
plt.bar(rating_counts.index, rating_counts)
plt.xlabel("Ratings")
plt.ylabel("Nombre d'oeuvres")
plt.title("Répartition du nombre d'oeuvres par rating")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
threshold = 0.01  # Par exemple, 1%

# Calculer le pourcentage de chaque catégorie
percentage = rating_counts / rating_counts.sum()

# Créer une nouvelle série où les petites catégories sont regroupées en 'Autres'
mask = percentage > threshold
rating_counts_grouped = rating_counts[mask]
rating_counts_grouped['Autres'] = rating_counts[~mask].sum()

# Afficher le graphique en camembert
plt.figure(figsize=(20, 10))
plt.pie(x=rating_counts_grouped, labels=rating_counts_grouped.index, autopct="%.2f%%")
plt.title("Répartition du nombre d'oeuvres par rating")
plt.show()
```

La **majorité des programmes proposés sur la plateforme sont en TV-MA**, un contenu qui n'est pas approprié aux moins de 17 ans, en raison de violence et de langage particulièrement grossier.

**Suivi par TV-14**, un contenu qui n'est pas approprié aux moins de 14 ans pour les mêmes raisons.

Lui même **suivi par TV-PG**, qui contiennent des éléments innapropriés pour les jeunes enfants.

Lui même **suivi par le R**, qui signifie que les moins de 17 ans doivent être accompagné d'un adulte.

Donc **une grande majorité du contenu de Netflix est violent,grossier ou inadapté à un jeune public.**

**Plus 75% du contenu de la plateforme est innaproprié pour les enfants.**

#### e. La répartition de la durée des films du dataset

#### f. La répartition de la durée des séries du dataset

#### i. Le top 5 des films les plus longs

Le **film le plus long** présent sur la plateforme est **Black Mirror: Bandersnatch avec plus de 300 minutes (5 heures).**

Il s'agit en réalité d'un **film interractif contenant plus de 5 heures de séquences filmées.**

Cependant, **un seul visionnage ne dure pas 5 heures.**

##### Nuage de mots des descriptions

```python
# On extrait le texte de la description
text = " ".join(df["description"].str.lower())

# On récupère les stop words
stop_words = set(stopwords.words('english'))

# On tokenize
tokens = word_tokenize(text)

# On filtre en retirant les stop words et la ponctuation
filtered_text = [word for word in tokens if word.isalpha() and word not in stop_words]
```

```python
# On calcule la fréquence de distibution des mots
word_freq = nltk.FreqDist(filtered_text)

# On récupère la distribution des 20 mots les plus fréquents
top_20 = word_freq.most_common(20)
top_20_dict = dict(top_20)

# On l'affiche la distribution des 20 mots les plus fréquents
plt.figure(figsize=(24, 6))
plt.bar(top_20_dict.keys(), top_20_dict.values())
plt.xlabel("Mots")
plt.ylabel("Fréquence")
plt.title("Top 20 des mots les plus fréquents dans les descriptions")
plt.show()
```

```python
# On crée un nuage de mots
wordcloud = WordCloud(width=800, height=200, background_color='white').generate_from_frequencies(word_freq)

# On l'affiche
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

On observe que les **mots les plus présents** dans les descriptions sont **"life", "young", "new" et "family".**

Ces mots sont souvent utilisés dans les descriptions sur Netflix pour plusieurs raisons:

- **Reflète le contenu**: Ces mots peuvent refléter les thèmes communs et les sujets abordés dans les films et les séries télévisées sur Netflix. Par exemple, de nombreux spectacles peuvent tourner autour de la “vie” quotidienne, des défis de la “jeunesse”, des “nouvelles” aventures ou des dynamiques “familiales”.

- **Attirer un large public**: Ces mots sont assez universels et peuvent attirer un large éventail de téléspectateurs. Par exemple, presque tout le monde peut se rapporter à des thèmes de “vie” ou de “famille”, et les mots comme “jeune” et “nouveau” peuvent susciter un sentiment d’excitation ou de nouveauté.

- **Stratégies basées sur les données**: Netflix utilise des stratégies basées sur les données pour décider du contenu à produire et à promouvoir. Ils analysent les mots les plus couramment utilisés dans les titres et les descriptions pour comprendre les thèmes et les sujets qui résonnent le plus avec leur public.
